# merge dsets

```python
import pandas as pd
```

```python
gap = pd.read_csv('gapminder.csv')
gdp = pd.read_csv('gdp2.csv')
```

```python
gap
```

```python
new = gap.merge(gdp, left_on=['country','Year'], right_on=['name', 'time'], how='inner')
```

```python
new = new.drop(columns=['Unnamed: 0', 'GDP', 'name', 'time'])
new
```

```python
new.to_csv('gapminder.csv')
```
