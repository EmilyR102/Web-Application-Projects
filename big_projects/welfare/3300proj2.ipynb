```python
import numpy as np
import pandas as pd
```

```python
gdp = pd.read_csv("gapminder_gdp.csv");
exp = pd.read_csv("gapminder_expectancy.csv");
pop = pd.read_csv("gapminder_population.csv");
```

```python
#wide --> tidy
gdp = pd.melt(gdp, id_vars = ['country'], var_name = 'Year', value_name = 'GDP')
exp = pd.melt(exp, id_vars = ['country'], var_name = 'Year', value_name = 'Life-Expectancy')
pop = pd.melt(pop, id_vars = ['country'], var_name = 'Year', value_name = 'Population')

print(gdp.head(n=5))
```

```python
gdp.sort_values(by=['country', 'Year']).groupby(['country'])
exp.sort_values(by=['country', 'Year']).groupby(['country'])
pop.sort_values(by=['country', 'Year']).groupby(['country'])

print(gdp.head(n=5))
print(exp.head(n=5))
print(pop.head(n=5))
```

```python
gapminder = gdp.merge(exp, how='left', on=['country', 'Year'])

display(gapminder)
```

```python
gapminder_full_df = gapminder.merge(pop, how='left', on=['country', 'Year'])

display(gapminder_full_df)
```

```python
gapminder_full = gapminder_full_df.sort_values(by=['country', 'Year'])

gapminder_full.head(n=5)

print(type(gapminder_full))
```

```python
gapminder_full.reset_index().to_csv('gapminder.csv')
```
