# fix data

```python
import pandas as pd
```

```python
gap = pd.read_csv('gapminder.csv')
```

```python
new_p=[]
for p in gap["Population"]:
    char = p[-1]
    if char == "M":
        new_p.append(float(p[:-1])*1000000)
    elif char == "k":
        new_p.append(float(p[:-1])*1000)
    elif char == "B":
        new_p.append(float(p[:-1])*1000000000)
    else:
        new_p.append(float(p))
gap['new_pop']=new_p
```
